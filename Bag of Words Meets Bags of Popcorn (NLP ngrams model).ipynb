# Preprocessing 
## Removing Stop_Words, Spaces and Punctuations

```python
nlp=spacy.load('en_core_web_sm')

def preprocess(text):
    doc=nlp(text)
    processed_text=[token.text for token in doc if not token.is_stop or token.is_space or token.is_punct]
    return " ".join(processed_text)
```

```python
df['new_review']=df['review'].apply(preprocess)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

# spliting the data into training and testing 
X_train,X_test,y_train,y_test=train_test_split(df['new_review'],df['sentiment'],train_size=0.8,random_state=22)
```

```python
from sklearn.pipeline import Pipeline
```

# Pipeline for Vectorizing and Training

```python
#creating a Pipeline for Vectorizing and training the model
clf = Pipeline([
    ('vectorizer', CountVectorizer(ngram_range=(1,2))),  
    # Using Bag of ngram model for more accurate result.
     ('Multi NB', MultinomialNB())   
    # navie bayes classifier for classification.
])
```

```python
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

#prints Detailed report
print(classification_report(y_test, y_pred))
```
